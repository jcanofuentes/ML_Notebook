```python
import tensorflow as tf 
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import sys
from tensorflow.python.client import device_lib

print("Python version: ", sys.version)
print("Tensorflow version: ", tf.__version__)
device_lib.list_local_devices()
```

Cargamos uno de los datasets que vienen con TF: Fashion-MNISt. Est치 compuesto de 60000 im치genes en B/N de 28x28px, clasificadas en 9 categorias. Adem치s incluye 10000 im치genes para pruebas.

```python
fashion_mnist = keras.datasets.fashion_mnist

(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
print( "Num train images :", x_train.shape)
print( "Num train labels :", len(y_train))
print(y_train[1:10])
print( "Num test images :", x_test.shape)
print( "Num test labels :", y_test.shape)
print(y_test[1:10])
```

Cambiamos el rango del valor de los pixels.

```python
print("Range of pixel values:", np.amin(x_train[0]), "-", np.amax(x_train[0]))

plt.figure()
plt.imshow(x_train[0],vmin=0,vmax=255)
plt.colorbar()
plt.grid(False)
plt.show()

x_train = x_train/255
x_test = x_test/255

print("Range of pixel values:", np.amin(x_train[0]), "-", np.amax(x_train[0]))
```

En el dataset, los valores de categoria se representan con una valor entero que va de 0 a 9. Creamos una lista con los nombres.

```python
class_names = ["Camiseta", "Pantal칩n","Jersey","Vestido","Abrigo","Sandalia","Camisa","Sneaker","Bolso","Ankle boot"]
```

Vamos a dibujar parte del dataset

```python
from mpl_toolkits.axes_grid1 import ImageGrid

fig = plt.figure(figsize=(8., 8.),frameon=False)

grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(4, 4),  # creates 4x4 grid of axes
                 axes_pad=0.5,  # pad between axes in inch.
                 share_all=True)

grid[0].get_yaxis().set_ticks([])
grid[0].get_xaxis().set_ticks([])

i = 0
for ax, im in zip(grid, x_train[0:16]):
    # Iterating over the grid returns the Axes.
    ax.imshow(im)
    label = class_names[y_train[i]]
    ax.text(0, 32, label, style='italic', bbox={'facecolor': 'green', 'alpha': 0.5, 'pad': 2})
    i = i+1
```

**Construimos el modelo**

1. La primera capa tan s칩lo formatea el conjunto de datos: pasamos del array 2d a un array 1d (con 28x28 = 784 px).
2. La segunda capa es densa, extrae  caracter칤sticas
3. La tercera capa es de Dropout. Desactiva aleatoriamente parte de las neuronas de cara a forwardpropagation y backwardpropagation, por lo que nos ayuda a evitar el overfitting 
4. La cuarta capa tambi칠n es densa, y obtiene como salida un array de 10 probabilidades que suman 1.   

```python
input = tf.keras.layers.Flatten(input_shape=(28,28))
hidden = tf.keras.layers.Dense(units=128,activation='relu')
hidden2 = tf.keras.layers.Dropout(0.2)
output = tf.keras.layers.Dense(units=10,activation='softmax') # Test with sigmoid

model = tf.keras.Sequential([input,hidden,hidden2,output])
```

**Compilamos el modelo**

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

**Entrenamos el modelo**

```python
model.fit(x_train, y_train, epochs=10)
```

**Evaluamos exactitud**

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)

print('Test loss:', test_loss)
print('Test accuracy:', test_acc)
```
